```python
import matplotlib.pyplot as plt
import os
import pandas as pd
plt.style.use('default')
import sys
sys.path.append('../')
from configs.config import get_config
```

```python
config_path = "/home/wwd/deeplearning/configs/config.yaml"
config = get_config(config_path)
```

```python
path = config.train.output + '/' + config.model.name
file = 'log.csv'

data = os.path.join(path, file)
df = pd.read_csv(data)
```

```python
# 参数设置优化
import matplotlib.ticker as ticker
line_width = 1.5
marker_style = 'o'
marker_size = 4
font_size = 14
legend_font_size = 14  # 放大图例字体
spine_width = 1.2      # 边框线条略粗

def style_subplot(ax, title, ylabel):
    ax.set_title(title, fontsize=font_size + 2)
    ax.set_ylabel(ylabel, fontsize=font_size)
    ax.tick_params(labelsize=font_size)
    ax.grid(False)

    # 仅保留左侧和下方边框
    for spine in ['top', 'right']:
        ax.spines[spine].set_visible(False)
    for spine in ['bottom', 'left']:
        ax.spines[spine].set_visible(True)
        ax.spines[spine].set_color('black')
        ax.spines[spine].set_linewidth(spine_width)

# 绘制图1：Train vs Val Loss
fig1, ax1 = plt.subplots(figsize=(10, 6))
ax1.plot(df['epoch'], df['train_loss'], label='Train Loss', 
         marker=marker_style, markersize=marker_size, linewidth=line_width, color='#1f77b4')
ax1.plot(df['epoch'], df['val_loss'], label='Val Loss', 
         marker=marker_style, markersize=marker_size, linewidth=line_width, color='#ff7f0e')

ax1.set_ylabel('Loss Value', fontsize=font_size)  # 设置纵轴标题为“损失函数值”
ax1.set_xlabel('Epoch', fontsize=font_size)
ax1.legend(fontsize=legend_font_size, loc='upper right')  # 图例字体更大
ax1.yaxis.set_major_locator(ticker.MaxNLocator(nbins=5))  # 最多5个主刻度，间距自动变稀疏
plt.tight_layout()
plt.show()

# --------------------------------------------------
# 图 2: MSE
# --------------------------------------------------
fig2, ax2 = plt.subplots(figsize=(12, 8))
ax2.plot(df['epoch'], df['MSE'], 
         marker=marker_style, linewidth=line_width, color='#2ca02c')
style_subplot(ax2, 'Mean Squared Error (MSE)', 'MSE')
ax2.set_xlabel('Epoch', fontsize=font_size)
plt.tight_layout()
plt.show()

# --------------------------------------------------
# 图 3: PSNR
# --------------------------------------------------
fig3, ax3 = plt.subplots(figsize=(12, 8))
ax3.plot(df['epoch'], df['PSNR'], 
         marker=marker_style, linewidth=line_width, color='#d62728')
style_subplot(ax3, 'PSNR', 'PSNR (dB)')
ax3.set_xlabel('Epoch', fontsize=font_size)
plt.tight_layout()
plt.show()

# --------------------------------------------------
# 图 4: SNR
# --------------------------------------------------
fig4, ax4 = plt.subplots(figsize=(12, 8))
ax4.plot(df['epoch'], df['SNR'], 
         marker=marker_style, linewidth=line_width, color='#9467bd')
style_subplot(ax4, 'SNR', 'SNR (dB)')
ax4.set_xlabel('Epoch', fontsize=font_size)
plt.tight_layout()
plt.show()

# --------------------------------------------------
# 图 5: SSIM
# --------------------------------------------------
fig5, ax5 = plt.subplots(figsize=(12, 8))
ax5.plot(df['epoch'], df['SSIM'], 
         marker=marker_style, linewidth=line_width, color='#8c564b')
style_subplot(ax5, 'SSIM', 'SSIM')
ax5.set_xlabel('Epoch', fontsize=font_size)
plt.tight_layout()
plt.show()

fig6, ax6 = plt.subplots(figsize=(12, 8))
ax6.plot(df['epoch'], df['lr'], 
         marker=marker_style, linewidth=line_width, color='#1f77b4')
style_subplot(ax6, 'lr', 'lr')
ax6.set_xlabel('Epoch', fontsize=font_size)
plt.tight_layout()
plt.show()
```
