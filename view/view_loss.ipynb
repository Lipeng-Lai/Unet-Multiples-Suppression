```python
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('default')
```

```python
# 读取CSV文件（假设文件名为loss_data.csv），仅加载前三列
df = pd.read_csv('../logs/training_log_20250318_225645.csv', usecols=['epoch', 'train_loss', 'val_loss'])

# 创建画布和坐标轴
plt.figure(figsize=(10, 6))

# 绘制训练损失曲线
plt.plot(df['epoch'], df['train_loss'], 
         label='Train Loss', 
         color='blue', 
         marker='o', 
         linestyle='-')

# 绘制验证损失曲线
plt.plot(df['epoch'], df['val_loss'], 
         label='Validation Loss', 
         color='red', 
         marker='s', 
         linestyle='--')

# 添加标签和标题
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.title('Training & Validation Loss Curve', fontsize=14, pad=20)

# 添加图例
plt.legend()

# 显示图表
plt.tight_layout()
plt.show()
```
